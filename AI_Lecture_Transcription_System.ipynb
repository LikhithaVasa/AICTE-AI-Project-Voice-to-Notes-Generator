```python
# ================================
# AI Lecture Voice-to-Notes Generator
# ================================

import yt_dlp
import whisper
import os

# ===== 1. Take YouTube URL Input =====
youtube_url = input("Enter YouTube URL: ")

# ===== 2. Download Audio from YouTube =====
audio_filename = "lecture_audio"

ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
    'outtmpl': audio_filename,
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([youtube_url])

audio_file = audio_filename + ".mp3"
print("вң… Audio downloaded successfully!")

# ===== 3. Load Whisper Tiny Model =====
print("вҸі Loading Whisper model...")
model = whisper.load_model("tiny")

# ===== 4. Transcribe Audio =====
print("вҸі Transcribing audio...")
result = model.transcribe(audio_file)
transcript_text = result["text"]

# Save Transcript
with open("Transcript.txt", "w", encoding="utf-8") as f:
    f.write(transcript_text)

print("вң… Transcript saved as Transcript.txt")

# ===== 5. Generate Study Notes =====
sentences = transcript_text.split(". ")
study_notes = ". ".join(sentences[:8])  # First 8 sentences as summary

with open("Study_Notes.txt", "w", encoding="utf-8") as f:
    f.write(study_notes)

print("вң… Study notes saved as Study_Notes.txt")

# ===== 6. Generate Simple Quiz =====
quiz = ""

for i in range(min(5, len(sentences))):
    quiz += f"Q{i+1}: {sentences[i]}?\n"
    quiz += "Answer: ______________________\n\n"

with open("Quiz.txt", "w", encoding="utf-8") as f:
    f.write(quiz)

print("вң… Quiz saved as Quiz.txt")

print("\nрҹҺү Process Completed Successfully!")
```
